```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range of x and y
x = np.linspace(0, 10, 100)  # Example x range
y = np.arange(0, 11, 0.5)  # Example y positions with 0.5 intervals

# Create a meshgrid for x and y
X, Y = np.meshgrid(x, y)

# Compute the Gaussian values for each x and y combination
max_gaussian = np.exp(-0.5 * ((y - 5) / 2)**2)  # Gaussian distribution for maximum values
gaussian = max_gaussian[:, np.newaxis] * np.exp(-0.5 * ((X - 5) / 2)**2)  # Gaussian lines with maximum values

# Find the indices of the maximum points along the x-axis
max_indices = np.argmax(gaussian, axis=1)

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Set a single color for all the lines and points
line_color = 'orange'
point_color = 'red'

# Plot the Gaussian lines
for yi in range(len(y)):
    ax.plot(x, y[yi] * np.ones_like(x), gaussian[yi], color=line_color, label=f"y = {y[yi]:.1f}")

# Plot red points at the maximum points
for yi, max_index in enumerate(max_indices):
    ax.scatter(x[max_index], y[yi], gaussian[yi, max_index], color=point_color)

# Remove the ticks
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# Customize the plot
ax.set_xlabel('X', fontsize=18)
ax.set_ylabel('Y', fontsize=18)
ax.set_zlabel('Z')

# Add z-axis label manually
ax.text2D(0.4, 0.95, 'dR/R Value', transform=ax.transAxes, fontsize=18)

# Set plot title
#ax.set_title('3D Plot with Gaussian Lines')

# Show the legend
# ax.legend()

# Show the plot
plt.show()
```
